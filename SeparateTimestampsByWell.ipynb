```python
#insert the name of all the files to separate time_stamps based on criteria 

import csv
# Function to separate time stamp based on criteria using files output from Creationtime script
def separate_data_by_wells(list):

    # Ask user to input the selected wells to search, separated by spaces 
    selected_wells = input("enter the selected wells to separate based on the category just printed:")
    
    # Initialize a dictionary to store time data 
    time_data = {}

    # Iterate over each file in the list 
    for file in list: 
        # Open the file for reading 
        with open(file,"r") as fh:
            # Read file as a dictionary 
            rows = csv.DictReader(fh)
            
            # Split the selected wells entered by the user 
            wells = selected_wells.split()

            # Extract unique identifier for the file 
            unique_file_identifier = file

            # Iterase over each row in the file 
            for row in rows:
                # Extract relevant information from the row 
                filename = row["filename"][-12:-4] #Extract the filename including field + channel
                exposure_time = row["time_per_well"][8:] #Extract exposure time 
                passage = row["passage_number"] # Extract passage number
                coordinates = row["coordinates"] # Extract coordinates 

                # Check if the filename starts with any of the selected wells, is the first field in each wells and exposure time is not 0   
                for n in wells:
                    if filename.startswith(n) and "f00" in filename and exposure_time != None:
                        # Create a unique identifier for the time data
                        unique_identifier = coordinates + "_" + passage + "_" + unique_file_identifier
                        # Store the exposure time with the unique identifier in the time_data dictionary
                        time_data[unique_identifier] = exposure_time
    
    # Return the dictionary containing time data                    
    return time_data
    
```

```python
# List all the separate files from Creationtime script
files = #insert file path containing timestamp data outputted from extractimagemetadata script

# list of categories that the timestamps data will be separated into
header = ["wells_with_beads","wells_before_beads","wells_adjacent_tobeads","wells_2nd_adjacent_to_beads"]
# Dictionary to store data for each category
Lists_combined = {}

# Iterate over each category in the header list
for entry in header:
    print("currently working on:", entry)

    # Store data for the current category using separate_data_by_wells function 
    Lists_combined[entry] = separate_data_by_wells(files)

    # Print completion message for current catgeory
    print("completed list:",entry)
```

```python
import csv
name_of_output_file = "output1" #name of the output file

with open(name_of_output_file + ".csv","w",newline='') as fh:
    #Define the fieldnames for the CSV file
    fieldnames = ["filename","time"]

    #Create CSV writer object
    rows = csv.DictWriter(fh,fieldnames = fieldnames)

    #Write the header row in the CSV file
    rows.writeheader()
    
    # Write an empty row
    rows.writerow({"filename":"","time":""}) 

    # Loop through each category in Lists_combined 
    for key,value in Lists_combined.items():

        # Write the category name as filename and an empty time 
        rows.writerow({"filename":key,"time":""})

        # Loop over each filename and time in the category
        for filename,time in value.items():

            #Write filename and corresponding timestamp to each line 
            rows.writerow({"filename":filename,"time":time})
        
        # Write an empty row to separate different categories 
        rows.writerow({"filename":"","time":""}) 
```
