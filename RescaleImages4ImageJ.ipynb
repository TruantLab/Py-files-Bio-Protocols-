```python
import cv2 as cv 
import os 
import numpy as np

# Function: open directory and output list of files in directory 
def get_files(get_directory = input("enter directory for input of images: ")):
  list_files = []
  for image in os.listdir(get_directory):
      file_name = get_directory + "/" + image #creates full path to each image in the directory 
      if file_name.endswith(".png") or file_name.endswith(".TIF") or file_name.endswith(".jpg") or file_name.endswith(".tif"): #optional: use to select images, if not remove 
        list_files.append(file_name)

  #print(list_files)
  return list_files

# Function to return the mean pixel value of the entire image set
def find_mean_pixel_value(list_files):
    #finds the mean pixel value for each image 
    number_pixels = 0
    pixel_values = 0
    mean_pixel_value = 0
    
    for file in list_files:
        img = cv.imread(file)
        #for each image in the list of files, computes the min and max pixel values 
        if img is not None:
            width, height,channels = img.shape

            #compute sum of pixel values in the image and total number of pixels
            pixel_values+=np.sum(img)

            #computes the number of pixels 
            number_pixels+=width*height

            # computes the minimum and maximum pixel values for each image 
            #image_min = np.min(img)
            #image_max = np.max(img)

    #computes the mean pixel values for the entire image set    
    mean_pixel_value = pixel_values/number_pixels
    return mean_pixel_value 

# Function to rescale each image based on the mean pixel value of the entire imageset
def rescale_images(mean_pixel_value,list_of_files):
   new_directory = input("enter the new directory path")
   for image in list_of_files:
      img = cv.imread(image) #read in each image 
      
      #find mean pixel value of image
      total_pixel_value = np.sum(img) # calculate the sum of all pixel value 
      width,height,channels = img.shape # extract the dimensions of each image
      pixels = width*height # calculate the number of pixels in image 
      imagemeanvalue = total_pixel_value / pixels # calculate mean pixel value 
        #print(imagemeanvalue)

      #rescale the image based on the average pixel value of the entire image set if the mean pixel value of image is not equal to mean pixel value of entire set
      if imagemeanvalue != mean_pixel_value:
            scaling_factor = mean_pixel_value/ imagemeanvalue #calculate rescaling factor 
            #print(scaling_factor)
            adjusted_img = cv.convertScaleAbs(img, alpha=scaling_factor, beta=0) # Rescale the image using rescaling factor
            new_filename = os.path.splitext(image)[0] + ".adjust.jpeg" # Create new filename for the adjusted image
            cv.imwrite(new_filename, adjusted_img) # Write adjusted image to the directory 

files = get_files() 
mean = find_mean_pixel_value(files)
rescale_images(mean,files)
```
